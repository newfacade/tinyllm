# Chain of Thought Prompting

```python
from utils import ask
```

```python
system_prompt = """ 
    You are a helpful assistant.
"""

def get_qustion_answer(question: str) -> str:
    messages = [
        {"role":"system", "content":system_prompt},
        {"role":"user", "content":question},
    ]
    response = ask(messages)
    # 后处理返回的结果
    ans = response.choices[0].message.content
    print(ans)
    return ans
```

## Standard Prompting

```python
question1 = """Q: Roger has 5 tennis balls. He buys 2 more cans of tennis balls.
Each can has 3 tennis balls. How many tennis balls does he have now?
A: The answer is 11.
Q: The cafeteria had 23 apples.
If they used 20 to make lunch and bought 6 more, how many apples do they have?
A:"""

ans1 = get_qustion_answer(question1)
```

## Chain of Thought -few shot

在few-shot的基础上增加reason step。

```python
question2 = """Q: Roger has 5 tennis balls. He buys 2 more cans of tennis balls.
Each can has 3 tennis balls. How many tennis balls does he have now?
A: Roger started with 5 balls. 2 cans of 3 tennis balls
each is 6 tennis balls. 5 + 6 = 11. The answer is 11.
Q: The cafeteria had 23 apples.
If they used 20 to make lunch and bought 6 more, how many apples do they have?
A:"""

ans2 = get_qustion_answer(question2)
```

## Chain of Thought - Zero shot

Zero-shot COT prompting 在标准提示词最后加入"Let's think step by step"。

```python
question3 = """Q: Roger has 5 tennis balls. He buys 2 more cans of tennis balls.
Each can has 3 tennis balls. How many tennis balls does he have now?
A: The answer is 11.

Q: The cafeteria had 23 apples.
If they used 20 to make lunch and bought 6 more, how many apples do they have?
A: Let's think step by step."""

ans3 = get_qustion_answer(question3)
```

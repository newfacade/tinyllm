```python
import pandas as pd
import random
```

```python
df = pd.read_excel('2025_12_leetcode.xlsx')
df
```

```python
result = []
for i, item in df.iterrows():
    result.append(item.to_dict())
result[0]
```

```python
result[-1]['链接'].split('?envType=')[0]
```

```python
from collections import defaultdict

info = defaultdict(list)
for item in result:
    url = '/'.join(item['链接'].split('/')[: 5])
    if not url.startswith('https://leetcode.cn/problems/'):
        print(url)
    problem = url.split('/')[-1]
    assert problem and problem != 'leetcode'
    item['链接'] = url
    del item['完成耗时']
    info[problem].append(item)
```

```python
r = [v[0] for v in info.values()]
random.shuffle(r)
r[0]
```

```python
random.shuffle(r)
```

```python
df = pd.DataFrame(r)
df.to_excel('problems.xlsx', index=False)
```
